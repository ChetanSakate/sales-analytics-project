```python
# ==========================================
# IMPORT REQUIRED LIBRARIES
# ==========================================

import pandas as pd
import numpy as np

# Display Settings
pd.set_option("display.max_columns", None)
pd.set_option("display.width", 1000)

print("Libraries imported successfully.")
```

```python
# ==========================================
# CONFIGURATION
# ==========================================

DATASET_PATH = "sales_data.csv"

HIGH_VALUE_THRESHOLD = 30000

CURRENCY_SYMBOL = "₹"
```

```python
# ==========================================
# DATA LOADING FUNCTIONS
# ==========================================

def load_dataset(file_path):
    """
    Load CSV dataset into pandas dataframe.
    """

    df = pd.read_csv(file_path)

    print(f"Dataset loaded successfully.")
    print(f"Total Records: {len(df)}")

    return df
```

```python
# ==========================================
# DATA PREPARATION FUNCTIONS
# ==========================================

def prepare_date_features(df):
    """
    Convert date columns and create
    time intelligence features.
    """

    df["order_date"] = pd.to_datetime(df["order_date"])

    df["month"] = df["order_date"].dt.month

    df["month_name"] = df["order_date"].dt.month_name()

    df["year"] = df["order_date"].dt.year

    return df
```

```python
# ==========================================
# DATA QUALITY CHECK FUNCTIONS
# ==========================================

def check_null_values(df):
    """
    Check null values in dataset.
    """

    return df.isnull().sum()


def check_duplicates(df):
    """
    Check duplicate records.
    """

    return df.duplicated().sum()
```

```python
# ==========================================
# GENERIC AGGREGATION FUNCTION
# ==========================================

def aggregate_metrics(
    df,
    group_column,
    metric_column="revenue",
    agg_function="sum",
    ascending=False,
    top_n=None
):
    """
    Generic reusable aggregation function.
    """

    result = (
        df.groupby(group_column)[metric_column]
          .agg(agg_function)
          .sort_values(ascending=ascending)
          .reset_index()
    )

    if top_n:
        result = result.head(top_n)

    return result
```

```python
# ==========================================
# EXECUTIVE KPI FUNCTIONS
# ==========================================

def calculate_executive_kpis(df):
    """
    Calculate business KPIs.
    """

    kpis = {
        "total_revenue": df["revenue"].sum(),
        "total_orders": df["order_id"].nunique(),
        "total_customers": df["customer_name"].nunique(),
        "average_order_value": df["revenue"].mean(),
        "total_quantity_sold": df["quantity"].sum()
    }

    return kpis
```

```python
# ==========================================
# MONTHLY TREND FUNCTIONS
# ==========================================

def calculate_monthly_revenue(df):
    """
    Calculate monthly revenue trend.
    """

    monthly_revenue = (
        df.groupby("month")["revenue"]
          .sum()
          .reset_index()
    )

    monthly_revenue["growth_percent"] = (
        monthly_revenue["revenue"]
        .pct_change() * 100
    )

    return monthly_revenue
```

```python
# ==========================================
# FILTERING FUNCTIONS
# ==========================================

def filter_high_value_transactions(df, threshold):
    """
    Filter high-value transactions.
    """

    return df[df["revenue"] > threshold]


def filter_by_region(df, region):
    """
    Filter region-specific data.
    """

    return df[df["region"] == region]
```

```python
# ==========================================
# LOAD DATA
# ==========================================

df = load_dataset(DATASET_PATH)
```

```python
# ==========================================
# DATA PREVIEW
# ==========================================

df.head()
```

```python
# ==========================================
# DATASET INFORMATION
# ==========================================

df.info()
```

```python
# ==========================================
# DESCRIPTIVE STATISTICS
# ==========================================

df.describe()
```

```python
# ==========================================
# DATA PREPARATION
# ==========================================

df = prepare_date_features(df)

print("Date features prepared successfully.")
```

```python
# ==========================================
# NULL VALUE CHECK
# ==========================================

check_null_values(df)
```

```python
# ==========================================
# DUPLICATE CHECK
# ==========================================

check_duplicates(df)
```

```python
# ==========================================
# EXECUTIVE KPI SUMMARY
# ==========================================

kpis = calculate_executive_kpis(df)

print("========== EXECUTIVE KPI SUMMARY ==========")

print(f"Total Revenue       : {CURRENCY_SYMBOL}{kpis['total_revenue']:,.2f}")

print(f"Total Orders        : {kpis['total_orders']}")

print(f"Total Customers     : {kpis['total_customers']}")

print(f"Average Order Value : {CURRENCY_SYMBOL}{kpis['average_order_value']:,.2f}")

print(f"Total Quantity Sold : {kpis['total_quantity_sold']}")
```

```python
# ==========================================
# REGION PERFORMANCE
# ==========================================

region_revenue = aggregate_metrics(
    df=df,
    group_column="region"
)

region_revenue
```

```python
# ==========================================
# TOP CUSTOMER ANALYSIS
# ==========================================

top_customers = aggregate_metrics(
    df=df,
    group_column="customer_name",
    top_n=10
)

top_customers
```

```python
# ==========================================
# PRODUCT PERFORMANCE
# ==========================================

product_performance = aggregate_metrics(
    df=df,
    group_column="product"
)

product_performance
```

```python
# ==========================================
# CATEGORY PERFORMANCE
# ==========================================

category_performance = (
    df.groupby("category")
      .agg({
          "revenue": "sum",
          "quantity": "sum"
      })
      .sort_values(by="revenue", ascending=False)
      .reset_index()
)

category_performance
```

```python
# ==========================================
# MONTHLY REVENUE TREND
# ==========================================

monthly_revenue = calculate_monthly_revenue(df)

monthly_revenue
```

```python
# ==========================================
# HIGH VALUE TRANSACTIONS
# ==========================================

high_value_transactions = filter_high_value_transactions(
    df,
    HIGH_VALUE_THRESHOLD
)

high_value_transactions.head()
```

```python
# ==========================================
# REGION DATASETS
# ==========================================

west_region = filter_by_region(df, "West")

east_region = filter_by_region(df, "East")

south_region = filter_by_region(df, "South")

north_region = filter_by_region(df, "North")

print("Region datasets created successfully.")
```

```python
# ==========================================
# CUSTOMER ORDER FREQUENCY
# ==========================================

customer_orders = aggregate_metrics(
    df=df,
    group_column="customer_name",
    metric_column="order_id",
    agg_function="count"
)

customer_orders.columns = [
    "customer_name",
    "total_orders"
]

customer_orders
```

```python
# ==========================================
# AVERAGE CUSTOMER REVENUE
# ==========================================

avg_customer_revenue = aggregate_metrics(
    df=df,
    group_column="customer_name",
    metric_column="revenue",
    agg_function="mean"
)

avg_customer_revenue
```

```python
# ==========================================
# FINAL BUSINESS SUMMARY
# ==========================================

top_region = region_revenue.iloc[0]

best_customer = top_customers.iloc[0]

best_product = product_performance.iloc[0]

print("========== EXECUTIVE BUSINESS SUMMARY ==========")

print(f"Total Revenue Generated : {CURRENCY_SYMBOL}{kpis['total_revenue']:,.2f}")

print(f"Top Performing Region   : {top_region['region']}")

print(f"Top Customer            : {best_customer['customer_name']}")

print(f"Best Product            : {best_product['product']}")

print(f"Total Orders            : {kpis['total_orders']}")

print("================================================")
```
